## Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.


```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
#Создаем таблицу, куда считаем время жизни для каждого пользователя.
ratings_avg_timestamp = ratings.groupby('userId').agg(['min', 'max'])['timestamp'].reset_index()
ratings_avg_timestamp.columns = ['userId', 'min_timestamp', 'max_timestamp']

#кладем значение в отдельную колонку
ratings_avg_timestamp['avg_timestamp'] = ratings_avg_timestamp['max_timestamp'] - ratings_avg_timestamp['min_timestamp']
ratings_avg_timestamp.head()
```

```python
#Создаем таблицу, где считаем количество пользователей, поставивших оценку больше 100 раз
ratings_grouped = ratings.groupby(['userId']).size().reset_index()
ratings_grouped.columns = ['userId', 'quantity']
ratings_filtered = ratings_grouped[ratings_grouped['quantity'] > 100]

#Соединяем две таблицы и находим среднее
total = ratings_filtered.join(ratings_avg_timestamp['avg_timestamp'], how='inner')
total['avg_timestamp'].mean()
```

## Задание 2

Дана статистика услуг перевозок клиентов компании по типам:

    rzd - железнодорожные перевозки
    auto - автомобильные перевозки
    air - воздушные перевозки
    client_base - адреса клиентов


Необходимо сформировать две таблицы:


- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
#Таблица 1 с тремя типами выручки для каждого client_id без указания адреса клиента
rzd.merge(auto,how='outer').merge(air,how='outer').head(10)
```

```python
#Таблица 2 по типам выручки с указанием адреса клиента
client_base.merge(rzd,how='outer').merge(auto,how='outer').merge(air,how='outer').head(10)
```
